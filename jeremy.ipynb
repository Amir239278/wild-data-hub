```python
import pandas as pd
df = pd.read_csv(r"dataset\df_agrégé.csv")
df2 = pd.read_csv(r"dataset\df_stat2.csv")
```

```python
# Suppr colonnes inutiles + renommer colonnes
df2_cleaned = df2[[
    'code_geo', 
    'libelle_geo', 
    'Population au dernier recensement 2021', 
    'Population projetée en 2070 2070', 
    'Médiane du revenu disponible par UC 2020']
]
df2_cleaned = df2_cleaned.rename(columns={
    'code_geo': 'Code_département',
    'libelle_geo': 'Libellé_département',
    'Population au dernier recensement 2021': 'Population_2021',
    'Population projetée en 2070 2070': 'Population_2070',
    'Médiane du revenu disponible par UC 2020': 'Revenu_médian_2020'}
)
```

```python
# Création nouvelle colonne évolution en pourcentage (arrondi à 2 décimales)
df2_cleaned['Evolution'] = df2_cleaned['Population_2070'] - df2_cleaned['Population_2021']
df2_cleaned['Evolution_pourcentage'] = ((df2_cleaned['Population_2070'] - df2_cleaned['Population_2021']) / df2_cleaned['Population_2021'] * 100).round(2)
```

```python
# Passe le colonne Code_département en integer (en remplaçant 2A et 2B par 201 et 202)
df2_cleaned['Code_département'] = df2_cleaned['Code_département'].replace({
    '2A': 201,
    '2B': 202
}).astype(int)
```

```python
df2_cleaned
```

```python
# Passe la colonne code_département de df en float
df['code_departement'] = df['code_departement'].replace({
    '2A': 201,
    '2B': 202
}).astype(int)
```

```python
# Supprime les colonnes de df
df = df.drop(columns=[
    'id_mutation',
    'nature_mutation',
    'date_mutation',
    'numero_disposition',
    'adresse_numero',
    'adresse_nom_voie',
    'adresse_code_voie',
    'code_postal',
    'code_commune',
    'nom_commune',
    'id_parcelle',
    'code_type_local',
    'longitude',
    'latitude'])
df
```

```python
# Joins df et df2_cleaned sur la colonne code_departement
df_joined = df.merge(df2_cleaned, left_on='code_departement', right_on='Code_département', how='outer')

# Supprime la colonne Code_département de df_joined
df_joined = df_joined.drop(columns=['Code_département'])

# Créer un datafrape où type_local = Maison et où type_loca = Appartement
df_appartement = df_joined[df_joined['type_local'] == 'Appartement']
df_maison = df_joined[df_joined['type_local'] == 'Maison']
```

```python
# Groupe df_appartement par code_departement et calcule la moyenne des prix
df_appartement_grouped = df_appartement.groupby('code_departement').agg({
    'valeur_fonciere': 'mean',
    'prix_m2': 'mean',    
    'Revenu_médian_2020': 'mean',
    'Population_2021': 'mean',
    'Population_2070': 'mean',
    'Evolution': 'mean',
    'Evolution_pourcentage': 'mean'}).reset_index().round(2)

# Groupe df_maison par code_departement et calcule la moyenne des prix
df_maison_grouped = df_maison.groupby('code_departement').agg({
    'valeur_fonciere': 'mean',
    'prix_m2': 'mean',
    'Revenu_médian_2020': 'mean',
    'Population_2021': 'mean',
    'Population_2070': 'mean',
    'Evolution': 'mean',
    'Evolution_pourcentage': 'mean'}).reset_index().round(2)
```

```python
df_maison_grouped
```

```python
df_ranked_app = df_appartement_grouped.copy()

# Classement des départements (1 = meilleur)
df_ranked_app['rang_revenu'] = df_ranked_app['Revenu_médian_2020'].rank(ascending=False) # Classement par revenu médian, rang 1 = revenu le plus élevé
df_ranked_app['rang_prix'] = df_ranked_app['prix_m2'].rank(ascending=True) # Classement par metre carré, rang 1 = prix le plus bas
df_ranked_app['rang_evolution'] = df_ranked_app['Evolution_pourcentage'].rank(ascending=False) # Classement par évolution de la population, rang 1 = plus forte croissance

# Moyenne des rangs (ou pondérée si tu veux)
df_ranked_app['note_attractivite_app'] = (
    df_ranked_app['rang_revenu'] +
    df_ranked_app['rang_prix'] +
    df_ranked_app['rang_evolution']
) / 3

# Trier les meilleurs départements
df_ranked_app_sorted = df_ranked_app.sort_values('note_attractivite_app')

# Afficher les 10 meilleurs
print(df_ranked_app_sorted[['code_departement', 'note_attractivite_app']].head(10))
```

```python
df_ranked_maison = df_maison_grouped.copy()

# Classement des départements (1 = meilleur)
df_ranked_maison['rang_revenu'] = df_ranked_maison['Revenu_médian_2020'].rank(ascending=False) # Classement par revenu médian, rang 1 = revenu le plus élevé
df_ranked_maison['rang_prix'] = df_ranked_maison['prix_m2'].rank(ascending=True) # Classement par prix metre carré, rang 1 = prix le plus bas
df_ranked_maison['rang_evolution'] = df_ranked_maison['Evolution_pourcentage'].rank(ascending=False) # Classement par évolution de la population, rang 1 = plus forte croissance

# Moyenne des rangs (ou pondérée si tu veux)
df_ranked_maison['note_attractivite_maison'] = (
    df_ranked_maison['rang_revenu'] +
    df_ranked_maison['rang_prix'] +
    df_ranked_maison['rang_evolution']
) / 3

# Trier les meilleurs départements
df_ranked_maison_sorted = df_ranked_maison.sort_values('note_attractivite_maison')

# Afficher les 10 meilleurs
print(df_ranked_maison_sorted[['code_departement', 'note_attractivite_maison']].head(10))
```

```python
# Fusionne df_ranked_maison_sorted et df_ranked_app_sorted sur la colonne code_departement
df_ranked_combined = df_ranked_maison_sorted.merge(
    df_ranked_app_sorted,
    on='code_departement')

print(df_ranked_combined[['code_departement', 'note_attractivite_maison', 'note_attractivite_app']])
```

```python
df_ranked_combined['note_attractivite_moyenne'] = (
    df_ranked_combined['note_attractivite_maison'] +
    df_ranked_combined['note_attractivite_app']
) / 2
```
