```python
df_departement = df_stat[df_stat['code_geo'].str.len() <= 2]
df_departement
```

```python
# recuperation des données de l'observatoire des territoires

df_obervatoire = pd.read_csv(r"dataset/observatoire des territoires_20260627.csv", sep=';',skiprows=[0, 1],low_memory=False)

```

```python
df_obervatoire
```

```python
df_obervatoire.columns
```

```python
# MERGE stat + observatoire

df_stat2 = pd.merge(df_departement, df_obervatoire, left_on='code_geo', right_on='Code', how='left')
```

```python
df_stat2.drop(columns=['Code', 'Libellé'], inplace=True)
```

```python
df_stat2.head()
```

```python
df_stat2.info()
```

```python
# convertir les colonnes en numérique

df_stat2["Médiane du revenu disponible par UC 2020" ] = pd.to_numeric(df_stat2["Médiane du revenu disponible par UC 2020"], errors='coerce')
df_stat2["Nombre de personnes en âge de travailler (15-64 ans) 2021" ] = pd.to_numeric(df_stat2["Nombre de personnes en âge de travailler (15-64 ans) 2021"], errors='coerce')
df_stat2["Nombre d'actifs de 15-64 ans 2021" ] = pd.to_numeric(df_stat2["Nombre d'actifs de 15-64 ans 2021"], errors='coerce')
df_stat2["Nombre de chômeurs de 15-64 ans 2021" ] = pd.to_numeric(df_stat2["Nombre de chômeurs de 15-64 ans 2021"], errors='coerce')
df_stat2["Population au dernier recensement 2021" ] = pd.to_numeric(df_stat2["Population au dernier recensement 2021"], errors='coerce')
```

```python
# calcul du taux de chômage
df_stat2['taux de chomage'] = round((df_stat2['Nombre de chômeurs de 15-64 ans 2021'] / df_stat2["Nombre d'actifs de 15-64 ans 2021"]) * 100,2)
```

```python
# Export du dataframe au format CSV
df_stat2.to_csv('dataset/df_stat2.csv', index=False)
```
